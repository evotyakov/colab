```python
# Подключение библиотек

import os
import re
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler #
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.base import BaseEstimator, ClassifierMixin
```

```python
# Загрузка файлов

def load_texts(folder_path, author):
  texts = []
  for filename in os.listdir(folder_path):
    if author in filename.lower():
      with open(os.path.join(folder_path, filename),
                'r', encoding='utf-8') as file:
        texts.append(file.read())
  return texts

zoshchenko_texts = load_texts('/content/drive/MyDrive/Colab Notebooks/',
                              'zoshchenko')
karamzin_texts = load_texts('/content/drive/MyDrive/Colab Notebooks/',
                         'karamzin')
```

```python
# Обработка и разбиение текстов на части

def clean_text(text):
  text = re.sub(r'[—–-](то|ка|что|за|нибудь)\b', '', text)
  text = re.sub(r'(\w+)-(\w+)', r'\1 \2', text)
  text = re.sub(r'\[.*?\]', '', text)
  text = re.sub(r'\b[IVXLCDM]+\.?\b', '', text, flags=re.IGNORECASE)
  text = re.sub(r'[^\w\s.!?]', '', text)
  text = re.sub(r'\.{3,}', '.', text)
  text = re.sub(r'\.{2,}', '', text)
  text = re.sub(r'[\r\n]+', ' ', text)
  text = re.sub(r'[ \t]+', ' ', text)
  text = text.replace('\xa0', '')
  text = text.strip()

  return text.lower()

def split_into_parts(text, target_length):
  sentences = re.split(r'(?<=[.!?])\s+', text)
  parts = []
  current_part = []
  current_length = 0

  for sent in sentences:
    sent_words = sent.split()
    if current_length + len(sent_words) <= target_length:
      current_part.append(sent)
      current_length += len(sent_words)
    else:
      if current_part:
        parts.append(' '.join(current_part))
      current_part = [sent]
      current_length = len(sent_words)

  if current_part:
      parts.append(' '.join(current_part))

  return parts

target_length = 1000

zoshchenko_parts = []
for text in zoshchenko_texts:
  cleaned = clean_text(text)
  zoshchenko_parts.extend(split_into_parts(cleaned, target_length))

karamzin_parts = []
for text in karamzin_texts:
  cleaned = clean_text(text)
  karamzin_parts.extend(split_into_parts(cleaned, target_length))

if zoshchenko_parts:
    zoshchenko_parts = zoshchenko_parts[:-1]

if karamzin_parts:
    karamzin_parts = karamzin_parts[:-1]

rd.seed(42)
zoshchenko_parts = rd.sample(zoshchenko_parts, 500)
karamzin_parts = rd.sample(karamzin_parts, 500)
```

```python
print(len(karamzin_parts), len(zoshchenko_parts))
```

```python
# Вычисление характеристик и создание датафрейма

def part_analysis(part):
  if not part.strip():
    return [0, 0]

  sentences = re.split(r'[.!?]+', part)
  sentences = [s.strip() for s in sentences if s.strip()]

  if not sentences:
    return [0, 0]

  total_s = len(sentences)
  total_w = 0
  total_c = 0

  for sentence in sentences:
    words = sentence.split()
    num_words = len(words)
    total_w += num_words
    total_c += sum(len(word) for word in words)

  avg_word_len = total_c / total_w if total_w else 0
  avg_sentence_len = total_w / total_s if total_s else 0

  return [avg_word_len, avg_sentence_len]

data = []

for part in zoshchenko_parts:
  features = part_analysis(part)
  data.append({'author': 'Зощенко',
               'avg_word_length': features[0],
               'avg_sent_length': features[1]})

for part in karamzin_parts:
  features = part_analysis(part)
  data.append({'author': 'Карамзин',
               'avg_word_length': features[0],
               'avg_sent_length': features[1]})

df = pd.DataFrame(data)
```

```python
# Построение графика

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,
                x='avg_word_length',
                y='avg_sent_length',
                hue='author',
                alpha=0.7)
plt.xlabel('Средняя длина слова')
plt.ylabel('Среднее число слов в предложении')
plt.title('Анализ текстов')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Разделение данных и классификация

df['author_label'] = df['author'].map({'Зощенко': 0, 'Карамзин': 1})

X = df[['avg_word_length', 'avg_sent_length']]
y = df['author_label']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=5)

# Модель 1 (Подбор коэффициентов)
class ManualClassifier:
    def __init__(self, coef, intercept):
        self.coef_ = coef
        self.intercept_ = intercept

    def predict(self, X):
        if hasattr(X, 'values'):
            X = X.values
        scores = X @ self.coef_.T + self.intercept_
        return (scores > 0).astype(int).flatten()

model1 = ManualClassifier(coef=np.array([[12, 1]]),
                          intercept=np.array([-83]))
y_pred1 = model1.predict(X_test)

# Модель 2 (Случайный поиск)
class RandomSearchClassifier:
  def __init__(self, n_iter=1000, coef_range=(-10, 10),
               intercept_range=(-100, 100)):
    self.n_iter = n_iter
    self.coef_range = coef_range
    self.intercept_range = intercept_range
    self.coef_ = None
    self.intercept_ = None

  def fit(self, X, y):
    if hasattr(X, 'values'):
        X = X.values
    if hasattr(y, 'values'):
        y = y.values

    best_acc = -1
    best_coef = None
    best_intercept = None

    for _ in range(self.n_iter):
      coef = np.random.uniform(self.coef_range[0],
                               self.coef_range[1],
                               size=X.shape[1])
      intercept = np.random.uniform(self.intercept_range[0],
                                    self.intercept_range[1])
      scores = X @ coef + intercept
      preds = (scores > 0).astype(int)
      acc = np.mean(preds == y)

      if acc > best_acc:
        best_acc = acc
        best_coef = coef
        best_intercept = intercept

      self.coef_ = best_coef.reshape(1, -1)
      self.intercept_ = best_intercept

  def predict(self, X):
    if hasattr(X, 'values'):
      X = X.values
    scores = X @ self.coef_.T + self.intercept_
    return (scores > 0).astype(int).flatten()

model2 = RandomSearchClassifier(n_iter=1000)
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)


# Модель 3 (LogisticRegression)
model3 = LogisticRegression()
model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_test)


# Точность и таблица классификации
print("Model 1 (Manual):")
print("Accuracy:", round(accuracy_score(y_test, y_pred1), 3))
print(confusion_matrix(y_test, y_pred1), "\n")

print("Model 2 (RandomSearch):")
print("Accuracy:", round(accuracy_score(y_test, y_pred2), 3))
print(confusion_matrix(y_test, y_pred2), "\n")

print("Model 3 (LogisticRegression):")
print("Accuracy:", round(accuracy_score(y_test, y_pred3), 3))
print(confusion_matrix(y_test, y_pred3))
```

```python
# Построение графика с разделяющими прямыми

df_test = X_test.copy()
df_test['author'] = y_test.map({0: 'Зощенко', 1: 'Карамзин'})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_test,
                x='avg_word_length',
                y='avg_sent_length',
                hue='author',
                alpha=0.7)

# Прямая, нарисованная вручную
x_min = X_test['avg_word_length'].min()
x_max = X_test['avg_word_length'].max()
x_values = np.linspace(x_min, x_max, 100)
y_manual = -12*x_values + 83
plt.plot(x_values, y_manual, 'm--', linewidth=2,
         label='Manual: 12x + y - 83 = 0')

# Прямая RandomSearch
if hasattr(model2, 'coef_'):
  w1 = model2.coef_[0][0]
  w2 = model2.coef_[0][1]
  b = model2.intercept_
  y_lr = (-w1*x_values - b)/w2
  plt.plot(x_values, y_lr, 'r-', linewidth=2,
           label=f'RandomSearch: {w1:.2f}x + {w2:.2f}y + {b:.2f} = 0')

# Прямая LogisticRegression
if hasattr(model3, 'coef_'):
  w1 = model3.coef_[0][0]
  w2 = model3.coef_[0][1]
  b = model3.intercept_[0]
  y_lr = (-w1*x_values - b)/w2
  plt.plot(x_values, y_lr, 'c-', linewidth=2,
           label=f'LogisticRegression: {w1:.2f}x + {w2:.2f}y + {b:.2f} = 0')

plt.xlabel('Средняя длина слова')
plt.ylabel('Среднее число слов в предложении')
plt.title('Сравнение разделяющих прямых на тестовой выборке')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Подсчёт точности классификации для частей разных размеров

result = []

for text in zoshchenko_texts:
  cleaned1 = clean_text(text)

for text in karamzin_texts:
  cleaned2 = clean_text(text)

for part_length in range(50, 1330, 20):
  zoshchenko_parts_new = []
  karamzin_parts_new = []

  zoshchenko_parts_new.extend(split_into_parts(cleaned1, part_length))
  karamzin_parts_new.extend(split_into_parts(cleaned2, part_length))

  rd.seed(42)
  zoshchenko_parts_new = rd.sample(zoshchenko_parts_new, 500)
  karamzin_parts_new = rd.sample(karamzin_parts_new, 500)

  data_new = []

  for part in zoshchenko_parts_new:
    features = part_analysis(part)
    data_new.append({'author': 'Зощенко',
                'avg_word_length': features[0],
                'avg_sent_length': features[1]})

  for part in karamzin_parts_new:
    features = part_analysis(part)
    data_new.append({'author': 'Карамзин',
                'avg_word_length': features[0],
                'avg_sent_length': features[1]})

  df_new = pd.DataFrame(data_new)

  df_new['author_label'] = df_new['author'].map({'Зощенко': 0,
                                                 'Карамзин': 1})

  X = df_new[['avg_word_length', 'avg_sent_length']]
  y = df_new['author_label']

  X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                      test_size=0.3,
                                                      random_state=42)

  model = LogisticRegression()
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  result.append({'part_length': part_length,
                 'accuracy': accuracy_score(y_test, y_pred)})

df_parts = pd.DataFrame(result)
```

```python
# График зависимости точности классификации от размера частей

plt.figure(figsize=(10, 6))
sns.regplot(data=df_parts,
            x='part_length',
            y='accuracy',
            scatter_kws={'alpha': 0.7},
            line_kws={'color': 'red'}
)
plt.xlabel('Размер частей текста')
plt.ylabel('Точность классификации')
plt.title('Зависимость точности классификации от размера частей текста')
plt.grid(True)
plt.show()
```
