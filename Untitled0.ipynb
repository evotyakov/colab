```python
# Подключение библиотек

import os
import re
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler #
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.base import BaseEstimator, ClassifierMixin
```

```python
# Загрузка файлов

def load_texts(folder_path, author):
  texts = []
  for filename in os.listdir(folder_path):
    if author in filename.lower():
      with open(os.path.join(folder_path, filename),
                'r', encoding='utf-8') as file:
        texts.append(file.read())
  return texts

zoshchenko_texts = load_texts('/content/drive/MyDrive/Colab Notebooks/',
                              'zoshchenko')
karamzin_texts = load_texts('/content/drive/MyDrive/Colab Notebooks/',
                         'karamzin')
```

```python
# Обработка и разбиение текстов на части

def clean_text(text):
  text = re.sub(r'[—–-](то|ка|что|за|нибудь)\b', '', text)
  text = re.sub(r'(\w+)-(\w+)', r'\1 \2', text)
  text = re.sub(r'\[.*?\]', '', text)
  text = re.sub(r'\b[IVXLCDM]+\.?\b', '', text, flags=re.IGNORECASE)
  text = re.sub(r'[^\w\s.!?]', '', text)
  text = re.sub(r'\.{3,}', '.', text)
  text = re.sub(r'\.{2,}', '', text)
  text = re.sub(r'[\r\n]+', ' ', text)
  text = re.sub(r'[ \t]+', ' ', text)
  text = text.replace('\xa0', '')
  text = text.strip()

  return text.lower()

def split_into_parts(text, target_length):
  sentences = re.split(r'(?<=[.!?])\s+', text)
  parts = []
  current_part = []
  current_length = 0

  for sent in sentences:
    sent_words = sent.split()
    if current_length + len(sent_words) <= target_length:
      current_part.append(sent)
      current_length += len(sent_words)
    else:
      if current_part:
        parts.append(' '.join(current_part))
      current_part = [sent]
      current_length = len(sent_words)

  if current_part:
      parts.append(' '.join(current_part))

  return parts

target_length = 1000

zoshchenko_parts = []
for text in zoshchenko_texts:
  cleaned = clean_text(text)
  zoshchenko_parts.extend(split_into_parts(cleaned, target_length))

karamzin_parts = []
for text in karamzin_texts:
  cleaned = clean_text(text)
  karamzin_parts.extend(split_into_parts(cleaned, target_length))

if zoshchenko_parts:
    zoshchenko_parts = zoshchenko_parts[:-1]

if karamzin_parts:
    karamzin_parts = karamzin_parts[:-1]

rd.seed(42)
zoshchenko_parts = rd.sample(zoshchenko_parts, 500)
karamzin_parts = rd.sample(karamzin_parts, 500)
```

```python
# Вычисление характеристик и создание датафрейма

def part_analysis(part):
  if not part.strip():
    return [0, 0]

  sentences = re.split(r'[.!?]+', part)
  sentences = [s.strip() for s in sentences if s.strip()]

  if not sentences:
    return [0, 0]

  total_s = len(sentences)
  total_w = 0
  total_c = 0

  for sentence in sentences:
    words = sentence.split()
    num_words = len(words)
    total_w += num_words
    total_c += sum(len(word) for word in words)

  avg_word_len = total_c / total_w if total_w else 0
  avg_sentence_len = total_w / total_s if total_s else 0

  return [avg_word_len, avg_sentence_len]

data = []

for part in zoshchenko_parts:
  features = part_analysis(part)
  data.append({'author': 'Зощенко',
               'avg_word_length': features[0],
               'avg_sent_length': features[1]})

for part in karamzin_parts:
  features = part_analysis(part)
  data.append({'author': 'Карамзин',
               'avg_word_length': features[0],
               'avg_sent_length': features[1]})

df = pd.DataFrame(data)
```

```python
# Построение графика

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,
                x='avg_word_length',
                y='avg_sent_length',
                hue='author',
                alpha=0.7)
plt.xlabel('Средняя длина слова')
plt.ylabel('Среднее число слов в предложении')
plt.title('Анализ текста')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Разделение данных и классификация

df['author_label'] = df['author'].map({'Zoshchenko': 0, 'Karamzin': 1})

X = df[['avg_word_length', 'avg_sent_length']]
y = df['author_label']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)

# Модель 1 (Подбор весов)
class ManualClassifier:
    def __init__(self, coef, intercept):
        self.coef_ = coef
        self.intercept_ = intercept

    def predict(self, X):
        if hasattr(X, 'values'):
            X = X.values
        scores = X @ self.coef_.T + self.intercept_
        return (scores > 0).astype(int).flatten()

model1 = ManualClassifier(coef=np.array([[12, 1]]),
                          intercept=np.array([-83]))

# Модель 2 (LogisticRegression)
model2 = LogisticRegression()
model2.fit(X_train, y_train)

y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)

# Точность и таблица классификации
print("Accuracy:", accuracy_score(y_test, y_pred1))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred1))

print("Accuracy:", accuracy_score(y_test, y_pred2))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred2))
```

```python
# Построение графика с разделяющими прямыми

df_test = X_test.copy()
df_test['author'] = y_test.map({0: 'Зощенко', 1: 'Карамзин'})

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_test,
                x='avg_word_length',
                y='avg_sent_length',
                hue='author',
                alpha=0.7)

# Прямая, нарисованная вручную
x_min = X_test['avg_word_length'].min()
x_max = X_test['avg_word_length'].max()
x_values = np.linspace(x_min, x_max, 100)
y_manual = -12*x_values + 83
plt.plot(x_values, y_manual, 'g--', linewidth=2,
         label='Ручная граница: 12x + y - 83 = 0')

# Прямая LogisticRegression
if hasattr(model2, 'coef_'):
    w1 = model2.coef_[0][0]
    w2 = model2.coef_[0][1]
    b = model2.intercept_[0]
    y_lr = (-w1*x_values - b)/w2
    plt.plot(x_values, y_lr, 'r-', linewidth=2,
             label=f'LR: {w1:.2f}x + {w2:.2f}y + {b:.2f} = 0')

plt.xlabel('Средняя длина слова')
plt.ylabel('Среднее число слов в предложении')
plt.title('Сравнение разделяющих прямых на тестовой выборке')
plt.legend()
plt.grid(True)

plt.show()
```
